# Merge data
Merge data from umsatzdaten_gekürzt.csv, kiwo.csv and wetter.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
print("✓ Imports successful!")
```

Read in the .csv-files.

```python
# Read in the CSV files
umsatz_df = pd.read_csv('../umsatzdaten_gekuerzt.csv')
kiwo_df = pd.read_csv('../kiwo.csv')
wetter_df = pd.read_csv('../wetter.csv')

# Display basic info about each dataset
print("Umsatzdaten shape:", umsatz_df.shape)
print("Kiwo shape:", kiwo_df.shape)
print("Wetter shape:", wetter_df.shape)
```

```python
print("Umsatz:",umsatz_df.columns)
print("Kiwo", kiwo_df.columns)
print("Wetter", wetter_df.columns)  
```

```python
merged_df = umsatz_df.merge(kiwo_df, on="Datum")\
    .merge(wetter_df, on="Datum")
    
print("Merged DataFrame shape:", merged_df.shape)    
print("Merged DataFrame:", merged_df.head())
```
